```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
movies = pd.read_csv('movies2004.csv')
```

```python
plot = movies['Plot'][27]
```

```python
# Removing punctuation
import string
nopunc = [char for char in plot if char not in string.punctuation]
nopunc = ''.join(nopunc)
```

```python
# Removing stopwords
import nltk
from nltk.corpus import stopwords
nostops = [word.lower() for word in nopunc.lower().split() if word not in stopwords.words('english')]
```

```python
# Using NLP library - spacy - to label words and pick persons
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp(plot)

pers_labels = [(w.text, w.label_) for w in doc.ents]

# Counter counts occurances of list items
from collections import Counter
key,label = zip(*Counter(pers_labels).keys())

key_label_count = pd.DataFrame.from_records([Counter(pers_labels).values(), key, label]).transpose()
key_label_count.rename(columns={0:'Count', 1:'Key', 2:'Label'}, inplace=True)
key_label_count.sort_values('Count', ascending=False, inplace=True)
key_label_count
```
