```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('dark')
%matplotlib inline
```

```python
df = pd.read_csv('imdb_movies.csv')
```

```python
df.head()
```

```python
df.release_date = pd.to_datetime(df.release_date)
```

```python
df['release_date'].dt.month.value_counts().sort_index().plot(kind='bar')
```

```python
df['release_date'].dt.dayofweek.value_counts().sort_index().plot(kind='bar')
```

```python
df.votes_number = df.votes_number.apply(lambda x: int(('').join(x.split(','))))
```

```python
plt.figure(figsize = (12,6))
sns.distplot(df.votes_number, kde=False)
```

```python
df.groupby(['year', 'votes_number']).size()
```
